Give the five number summary for the column `Age`?


**Double click here for hint**.

<!--
min,q1,median,q3,max of a column are its five number summary.

```python
# your code goes here


# Extract the 'Age' column
ages = df['Age']

# Calculate the five number summary
minimum = ages.min()
q1 = ages.quantile(0.25)
median = ages.median()
q3 = ages.quantile(0.75)
maximum = ages.max()

# Display the five number summary
print("Five Number Summary for Age:")
print(f"Minimum: {minimum}")
print(f"Q1 (First Quartile): {q1}")
print(f"Median (Second Quartile): {median}")
print(f"Q3 (Third Quartile): {q3}")
print(f"Maximum: {maximum}")
```

```python
# Calculate correlation coefficients between 'Age' and all other numerical columns
correlations = df.corr()['Age'].sort_values()

# The column with the most negative correlation with 'Age' (excluding 'Age' itself)
most_negative_corr_column = correlations.index[0]

print("Column with the most negative correlation with 'Age':", most_negative_corr_column)



# Calculate correlation coefficients between 'Age' and all other numerical columns
correlations = df.corr()['Age'].sort_values(ascending=False)

# The column with the highest correlation with 'Age' (excluding 'Age' itself)
highest_corr_column = correlations.index[1]  # Index 0 is 'Age' itself, so we take index 1

# The correlation coefficient value itself
highest_corr_value = correlations.iloc[1]

print("Column with the highest correlation with 'Age':", highest_corr_column)
print("Correlation coefficient value:", highest_corr_value)
```

## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
import matplotlib.pyplot as plt


# Filter out NaN values in ConvertedComp if any
df_filtered = df.dropna(subset=['ConvertedComp'])

# Plotting box plot to visualize outliers
plt.figure(figsize=(10, 6))
plt.boxplot(df_filtered['ConvertedComp'], vert=False, patch_artist=True, boxprops=dict(facecolor='skyblue'))

# Customize labels and title
plt.xlabel('Converted Compensation')
plt.title('Box Plot of Converted Compensation (Checking for Outliers)')

# Show plot
plt.grid(True)
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here

# Filter out NaN values in ConvertedComp if any
df_filtered = df.dropna(subset=['ConvertedComp'])

# Calculate quartiles
Q1 = df_filtered['ConvertedComp'].quantile(0.25)
Q3 = df_filtered['ConvertedComp'].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

print(f"The Interquartile Range (IQR) for ConvertedComp is: {IQR}")
```

Find out the upper and lower bounds.


```python
# your code goes here

# Filter out NaN values in ConvertedComp if any
df_filtered = df.dropna(subset=['ConvertedComp'])

# Calculate quartiles and IQR
Q1 = df_filtered['ConvertedComp'].quantile(0.25)
Q3 = df_filtered['ConvertedComp'].quantile(0.75)
IQR = Q3 - Q1

# Calculate lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Lower Bound for outliers: {lower_bound}")
print(f"Upper Bound for outliers: {upper_bound}")
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here


# Filter out NaN values in ConvertedComp if any
df_filtered = df.dropna(subset=['ConvertedComp'])

# Calculate quartiles and IQR
Q1 = df_filtered['ConvertedComp'].quantile(0.25)
Q3 = df_filtered['ConvertedComp'].quantile(0.75)
IQR = Q3 - Q1

# Calculate lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df_filtered[(df_filtered['ConvertedComp'] < lower_bound) | (df_filtered['ConvertedComp'] > upper_bound)]

# Count number of outliers
num_outliers = outliers.shape[0]

print(f"Number of outliers in ConvertedComp column: {num_outliers}")
```

```python
# Plot the boxplot for the 'Age' column
plt.figure(figsize=(10, 6))
plt.boxplot(df['Age'].dropna(), vert=False)
plt.title('Boxplot of Age')
plt.xlabel('Age')
plt.show()

# Calculate Q1 and IQR for 'Age'
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower bound for outliers
lower_bound = Q1 - 1.5 * IQR

# Identify the number of outliers below Q1
outliers_below_Q1 = df[df['Age'] < lower_bound]
num_outliers_below_Q1 = outliers_below_Q1.shape[0]

print("Number of outliers below Q1:", num_outliers_below_Q1)
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
# Filter out NaN values in ConvertedComp if any
df_filtered = df.dropna(subset=['ConvertedComp'])

# Calculate quartiles and IQR
Q1 = df_filtered['ConvertedComp'].quantile(0.25)
Q3 = df_filtered['ConvertedComp'].quantile(0.75)
IQR = Q3 - Q1

# Calculate lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Create new DataFrame without outliers
df_no_outliers = df_filtered[(df_filtered['ConvertedComp'] >= lower_bound) & (df_filtered['ConvertedComp'] <= upper_bound)]

# Optionally, reset index for the new DataFrame
df_no_outliers = df_no_outliers.reset_index(drop=True)

# Display or use df_no_outliers as needed
print("New DataFrame without outliers:")
print(df_no_outliers)
```

## Correlation


### Finding correlation


Find the correlation between `Age` and all other numerical columns.


```python
# your code goes here
# Filter out numerical columns excluding 'Age'
numerical_columns = df_no_outliers.select_dtypes(include=['int64', 'float64']).columns

# Calculate correlation between Age and other numerical columns
correlations = df_no_outliers[numerical_columns].corrwith(df_no_outliers['Age'])

# Display the correlations
print("Correlation between Age and other numerical columns:")
print(correlations)
```

```python
# Calculate the IQR for 'ConvertedComp'
Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bound for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove the outliers
df_no_outliers = df[(df['ConvertedComp'] >= lower_bound) & (df['ConvertedComp'] <= upper_bound)]

# Calculate the median of the 'ConvertedComp' column after removing outliers
median_converted_comp_no_outliers = df_no_outliers['ConvertedComp'].median()
print("Median of ConvertedComp after removing outliers:", median_converted_comp_no_outliers)
```

```python
# Calculate Q1, Q3, and IQR for the 'ConvertedComp' column
Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df_no_outliers = df[(df['ConvertedComp'] >= lower_bound) & (df['ConvertedComp'] <= upper_bound)]

# Calculate the mean of the 'ConvertedComp' column after removing outliers
mean_converted_comp_no_outliers = df_no_outliers['ConvertedComp'].mean()

print("Mean of ConvertedComp after removing outliers:", mean_converted_comp_no_outliers)
```

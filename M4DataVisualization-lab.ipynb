```python
import pandas as pd
import sqlite3

# Load the dataset into a DataFrame
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m1_survey_data.csv"
df = pd.read_csv(dataset_url)

# Create a connection to the SQLite database
conn = sqlite3.connect('m1_survey_data.sqlite')

# Write the DataFrame to an SQLite table named 'm1_survey_data'
df.to_sql('m1_survey_data', conn, if_exists='replace', index=False)


# Function to replace null values in categorical columns with mode
def replace_nulls_with_mode(column_name):
    conn.execute(f"""
    UPDATE m1_survey_data
    SET {column_name} = (
        SELECT {column_name}
        FROM (
            SELECT {column_name}, COUNT(*) as cnt
            FROM m1_survey_data
            WHERE {column_name} IS NOT NULL
            GROUP BY {column_name}
            ORDER BY cnt DESC
            LIMIT 1
        )
    )
    WHERE {column_name} IS NULL;
    """)

# Replace null values in categorical columns with mode
categorical_columns = ['Hobbyist', 'OpenSourcer', 'OpenSource', 'Sexuality', 'Ethnicity']  # Add other categorical columns as needed
for column in categorical_columns:
    replace_nulls_with_mode(column)

# Function to replace null values in numeric columns with median
def replace_nulls_with_median(column_name):
    conn.execute(f"""
    UPDATE m1_survey_data
    SET {column_name} = (
        SELECT AVG(temp.{column_name})
        FROM (
            SELECT {column_name}
            FROM m1_survey_data
            WHERE {column_name} IS NOT NULL
            ORDER BY {column_name}
            LIMIT 2 - (SELECT COUNT(*) FROM m1_survey_data WHERE {column_name} IS NOT NULL) % 2
            OFFSET (SELECT (COUNT(*) - 1) / 2 FROM m1_survey_data WHERE {column_name} IS NOT NULL)
        ) AS temp
    )
    WHERE {column_name} IS NULL;
    """)

# Replace null values in numeric columns with median
numeric_columns = ['ConvertedComp']  # Add other numeric columns as needed
for column in numeric_columns:
    replace_nulls_with_median(column)

```

```python
# Calculate Q1, Q3, and IQR in Python
q1 = df['ConvertedComp'].quantile(0.25)
q3 = df['ConvertedComp'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Remove outliers using the calculated bounds in SQL
conn.execute(f"""
DELETE FROM m1_survey_data
WHERE ConvertedComp < {lower_bound}
   OR ConvertedComp > {upper_bound};
""")

# Commit changes and close the connection
conn.commit()
conn.close()
```

```python
# Create a connection to the SQLite database
conn = sqlite3.connect('m1_survey_data.sqlite')
# Function to replace null values in categorical columns with mode
def replace_nulls_with_mode(column_name):
    conn.execute(f"""
    UPDATE m1_survey_data
    SET {column_name} = (
        SELECT {column_name}
        FROM (
            SELECT {column_name}, COUNT(*) as cnt
            FROM m1_survey_data
            WHERE {column_name} IS NOT NULL
            GROUP BY {column_name}
            ORDER BY cnt DESC
            LIMIT 1
        )
    )
    WHERE {column_name} IS NULL;
    """)

# Function to replace null values in numeric columns with median
def replace_nulls_with_median(column_name):
    conn.execute(f"""
    UPDATE m1_survey_data
    SET {column_name} = (
        SELECT AVG(temp.{column_name})
        FROM (
            SELECT {column_name}
            FROM m1_survey_data
            WHERE {column_name} IS NOT NULL
            ORDER BY {column_name}
            LIMIT 2 - (SELECT COUNT(*) FROM m1_survey_data WHERE {column_name} IS NOT NULL) % 2
            OFFSET (SELECT (COUNT(*) - 1) / 2 FROM m1_survey_data WHERE {column_name} IS NOT NULL)
        ) AS temp
    )
    WHERE {column_name} IS NULL;
    """)

# Replace null values in all columns appropriately
columns_to_replace = {
    'categorical': ['Hobbyist', 'OpenSourcer', 'OpenSource', 'Sexuality', 'Ethnicity', 'Dependents', 'SurveyLength', 'SurveyEase'],
    'numeric': ['ConvertedComp']
}

for column in columns_to_replace['categorical']:
    replace_nulls_with_mode(column)

for column in columns_to_replace['numeric']:
    replace_nulls_with_median(column)

# Calculate Q1, Q3, and IQR in Python
q1 = df['ConvertedComp'].quantile(0.25)
q3 = df['ConvertedComp'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Remove outliers using the calculated bounds in SQL
conn.execute(f"""
DELETE FROM m1_survey_data
WHERE ConvertedComp < {lower_bound}
   OR ConvertedComp > {upper_bound};
""")

# Commit changes
conn.commit()

# Read the cleaned data into a DataFrame
cleaned_df = pd.read_sql_query("SELECT * FROM m1_survey_data", conn)
```

### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Create a connection to the SQLite database
conn = sqlite3.connect('m1_survey_data.sqlite')

# Query the DatabaseDesireNextYear data from the SQLite database
query = """
SELECT DatabaseDesireNextYear
FROM m1_survey_data
WHERE DatabaseDesireNextYear IS NOT NULL
"""

# Execute the query and load the data into a DataFrame
database_data = pd.read_sql_query(query, conn)

# Close the database connection
conn.close()

# Split the column into individual databases and explode the list
database_data = database_data['DatabaseDesireNextYear'].str.split(';', expand=True).stack()
database_data = database_data.reset_index(level=1, drop=True)
database_data.name = 'Database'

# Count the occurrences of each database
database_counts = database_data.value_counts()

# Get the top 5 databases
top_5_databases = database_counts.head(5)

# Plot a pie chart of the top 5 databases using matplotlib
plt.figure(figsize=(10, 7))
plt.pie(top_5_databases, labels=top_5_databases.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(5)))
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Create a connection to the SQLite database
conn = sqlite3.connect('m1_survey_data.sqlite')

# Define the age group criteria
age_group = "30-34"  # Adjust as needed based on your dataset

# Query to calculate median WorkWeekHrs and CodeRevHrs for the specified age group
query = f"""
SELECT Age,
       AVG(WorkWeekHrs) AS MedianWorkWeekHrs,
       AVG(CodeRevHrs) AS MedianCodeRevHrs
FROM m1_survey_data  
WHERE Age >= 30 AND Age <= 35
GROUP BY Age
ORDER BY Age
"""

# Execute the query and load the data into a DataFrame
df = pd.read_sql_query(query, conn)

# Close the database connection
conn.close()

# Plotting the stacked bar chart
plt.figure(figsize=(10, 6))

# Plotting median WorkWeekHrs
plt.bar(df['Age'], df['MedianWorkWeekHrs'], label='Median WorkWeekHrs')

# Plotting median CodeRevHrs on top of WorkWeekHrs
plt.bar(df['Age'], df['MedianCodeRevHrs'], bottom=df['MedianWorkWeekHrs'], label='Median CodeRevHrs')

# Adding labels and title
plt.xlabel('Age Group')
plt.ylabel('Hours')
plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.legend()

# Show plot
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Create a connection to the SQLite database
conn = sqlite3.connect('m1_survey_data.sqlite')

# Query to calculate median ConvertedComp for the specified age range
query = """
SELECT Age,
       AVG(ConvertedComp) AS MedianConvertedComp
FROM m1_survey_data  
WHERE Age >= 45 AND Age <= 60
GROUP BY Age
ORDER BY Age
"""

# Execute the query and load the data into a DataFrame
df = pd.read_sql_query(query, conn)

# Close the database connection
conn.close()

# Plotting the line chart
plt.figure(figsize=(10, 6))

plt.plot(df['Age'], df['MedianConvertedComp'], marker='o', linestyle='-', color='b', label='Median ConvertedComp')

# Adding labels and title
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.title('Median ConvertedComp for Ages 45-60')
plt.legend()

# Show plot
plt.grid(True)
plt.show()

```
